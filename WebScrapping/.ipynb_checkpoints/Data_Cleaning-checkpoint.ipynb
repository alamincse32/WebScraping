```python
import requests 
import pandas as pd
import numpy as np
import re
import os
import time
import re
```

```python
# remove special characters from the text

def remove_special_characters(text):
    # Newlines (replaced with space to preserve cases like word1\nword2)
    text = re.sub(r"\n+", " ", text)
    # Remove resulting ' '
    text = text.strip()
    text = re.sub(r'\s\s+', ' ', text)

    # emails
    text = re.sub('\S*@\S*\s?', '', text)

    # > Quotes
    text = re.sub(r'\"?\\?&?gt;?', '', text)

    # Bullet points/asterisk (bold/italic)
    text = re.sub(r'\*', '', text)
    text = re.sub('&amp;#x200B;', '', text)

    # remove the [removed] from text
    text = text.replace('[removed]', '')

    # remove the [deleted] from text
    text = text.replace('[deleted]', '')

    # things in parantheses or brackets
    text = re.sub(r'[\[.*?\]\(.*?\)]', '', text)

    # remove URLS
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text)

    # Strikethrough
    text = re.sub('~', '', text)
    

    # Exclamation mark remove
    text = re.sub('!', '', text)

    # Spoiler, which is used with < less-than (Preserves the text)
    text = re.sub('&lt;', '', text)
    text = re.sub(r'!(.*?)!', r'\1', text)

    # Code, inline and block
    text = re.sub('`', '', text)

    # Superscript (Preserves the text)
    text = re.sub(r'\^\((.*?)\)', r'\1', text)

    # Table
    text = re.sub(r'\|', ' ', text)
    text = re.sub(':-', '', text)

    # Heading
    text = re.sub('#', '', text)
    
    text = re.sub('"', '', text)
    
    text = re.sub('\'', '', text)
    text = re.sub(',', '', text)
    text = re.sub('`', '', text)

    return text
```

```python
path = "Results"
word_token = ['robot', 'welders', 'robot welders', 'virtual', 'reality', 'virtual reality', 'automation', 'automated', 'research', 'design', 'development', 'testing', 'digital', 'technology', ' ag ', 'sustainable', 'intellectual', 'property', 'intellectual property', ' iot ', 'industry 4.0', 'industry 5.0', 'industry', 'team', 'learn', 'train', 'flexible', 'quality', 'resources', ' smb ', 'programming', 'advance technology', 'compliance', 'industrial', 'automotive', 'recreational', 'transportation', 'detection', 'cyber', 'physical', 'cyber physical', 'cloud', 'computing', 'cloud computing', ' big ', 'data', 'big data', 'artificial', 'intelligence', 'artificial intelligence','5g', '3d', '3d printing', 'printing', '4d', '4d printing', 'twin', 'digital twin', 'wearable', 'devices', 'wearable devices', ' ai ', 'communication']
print(len(word_token))
```

```python
def data_read(directory):
    print(directory)
    df = pd.read_csv(directory, encoding="utf-8")
    df = df.drop_duplicates()
    data = []
    n = df.shape[0]
    for i in range(n):
        row = df.loc[i].dropna().tolist()
        row = ' '.join(row[1:])
        data.append(row)
    return ' '.join(data)
```

```python
for i in range(71, 72, 1):
    data = data_read(os.path.join(path,str(i),'data.csv')).lower()
    data = remove_special_characters(data)
    dict_ = {}
    for token in word_token:
        dict_[token] = len(re.findall(token, data))
    df = pd.DataFrame(dict_.items(), columns=['Tokens', str(i)])
    df.to_csv(os.path.join(path,str(i),'token_count.csv'))
```
